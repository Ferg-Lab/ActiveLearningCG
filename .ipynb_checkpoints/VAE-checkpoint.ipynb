## Preprocessing Peptide Chemistries
Here, the set of MARTINI oligopeptides containing 3 amino acids ($20^3=8000$ unique peptides total) are encoded according to their topology. The geometry of each peptide is represented using an adjacency matrix capturing the connected structure of the MARTINI beads. The different types of MARTINI beads are seperately represented with a one-hot encoded array. These two structures are used to represent each peptide in the form of input amenable to deep learning.

```python
# Define an alphabet of amino acids
aminoAcidAlphabet = 'arndcqeghilkmfpstwyv'
aminoAcidList = list(aminoAcidAlphabet)

# Define the size of oligopeptides we want to model
WING_SIZE = 3
im_size = WING_SIZE*5
bead_size = WING_SIZE*25

# Number of points we want to use for training/validation
input_length = 8000
validationSize = 0

# Create all unique combinations of 3-amino acids
wings = [p for p in itertools.product(aminoAcidList,repeat=WING_SIZE)]
shuffle(wings)

# Generate each matrix and bead array from the `Processing` class.
matrix = list()
peptide = list()
beads = list()
for wing in wings[:input_length]:
    residueString = ''.join(wing)
    a = Processing(residueString)
    peptide.append([residueString,a.mols])
    matrix.append(a.scaledMatrix.reshape(im_size,im_size,1))
    beads.append(a.beadArray)

matrix = np.array(matrix,dtype=np.float32)
peptide = np.asarray(peptide)
beads = np.array(beads,dtype=np.float32)

#One-hot encode all bead types
beads = OneHotEncoder().fit_transform(beads).A.astype('float32')
```

### Visualizaiton
A plot of the bead adjacency matrix (top) and one-hot encoded bead type array (bottom) for an oligopeptide (WAF).

```python
def showPeptide(peptideString):
    ''''
    Helper function for ploting and visualizing peptide adjacency matrix and bead type array
    # Inputs:
        peptideString (string):
             string of length `WING_SIZE` of amino acids to plot
    # Returns:
        none
    
    # Notes:
        Generates a plot of the adjacency matrix (top) and one-hot encoded bead type array (bottom)
    '''
    peptideIndex = np.where(peptide[:,0]==peptideString)[0][0]
    matshow(matrix[peptideIndex].reshape(im_size,im_size))
    plt.xlabel(peptide[peptideIndex][0].upper(),fontsize=22,labelpad=20)
    matshow(beads[peptideIndex].reshape(1,-1))
showPeptide('waf')
```

# Variational Autoencoder
A VAE is created to generate a low-dimensional embedding from the adjacency matrix and composition array inputs for each oligopeptide. The adjacency is treated with 2 convolutional layers, while the bead type array is encoded with a multilayer perceptron. The features extracted from each encoding are then concatenated and used in generating a lower dimensional latent space. Two multilayer perceptrons act as the decoder part of the network, upsampling the embedding to separately reconstruct the adjacency matrix and composition array inputs. 

```python
# reparameterization trick
# instead of sampling from Q(z|X), sample eps = N(0,I)
# z = z_mean + sqrt(var)*eps
def sampling(args):
    """
    Reparameterization trick by sampling an isotropic unit Gaussian.
    # Arguments:
        args (tensor): mean and log of variance of Q(z|X)
    # Returns:
        z (tensor): sampled latent vector
    """

    z_mean, z_log_var = args
    batch = K.shape(z_mean)[0]
    dim = K.int_shape(z_mean)[1]
    # by default, random_normal has mean=0 and std=1.0
    epsilon = K.random_normal(shape=(batch, dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

# Define the size of the embedding
latent_dim = 3

# Convolutional encoding
inputs_image = Input(shape = (im_size,im_size,1,),name ="Adjacency_Matrix")
# Encoding layer 1
h1_image = Conv2D(16,kernel_size=(5,5),strides=(1,1))(inputs_image)
h1_image_normed = BatchNormalization(axis=-1)(h1_image)
h1_image_activation = Activation('relu')(h1_image_normed)
h1_image_out = MaxPooling2D(pool_size=(2,2),strides=(2,2))(h1_image_activation)
# Encoding layer 2
h2_image = Conv2D(32,kernel_size=(4,4),strides=(1,1))(h1_image_out)
h2_image_normed = BatchNormalization(axis=-1)(h2_image)
h2_image_activation = Activation('relu')(h2_image_normed)
h2_image_out = MaxPooling2D(pool_size=(2,2),strides=(2,2))(h2_image_activation)
image_encoded = Flatten()(h2_image_out)
# Bead encoding
inputs_bead = Input(shape = (bead_size,),name = "Bead_Array")
# Encoding layer 1 # input is 75
h1_bead = Dense(50,activation='relu')(inputs_bead)
h2_bead = Dense(25,activation='relu')(h1_bead)
bead_encoded = Dense(14,activation='relu')(h2_bead)
# Concatenate your outputs
encoded = Concatenate()([image_encoded,bead_encoded])

z_mean = Dense(latent_dim, name='z_mean')(encoded)
z_log_var = Dense(latent_dim, name='z_log_var')(encoded)

# use reparameterization trick to push the sampling out as input
# note that "output_shape" isn't necessary with the TensorFlow backend
z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])
encoder = Model([inputs_image,inputs_bead], [z_mean, z_log_var, z], name='encoder')

# Decoder
latent_inputs = Input(shape=(latent_dim,), name='z_sampling')

# Convolutional decoder
decode_image_1 = Dense(125,activation='relu')(latent_inputs)
decode_image_2 = Dense(im_size*im_size,activation='sigmoid')(decode_image_1)
decode_image_out = Reshape((im_size,im_size,1))(decode_image_2)
# Bead decoder
decode_bead_1 = Dense(14,activation='relu')(latent_inputs)
decode_bead_2 = Dense(25,activation='relu')(decode_bead_1)
decode_bead_3 = Dense(50,activation='relu')(decode_bead_2)
decode_bead_out = Dense(bead_size,activation='sigmoid')(decode_bead_3)

decoder = Model(latent_inputs, [decode_image_out,decode_bead_out], name='decoder')

outputs = decoder(encoder([inputs_image,inputs_bead])[2])
vae = Model([inputs_image,inputs_bead], outputs, name='vae')

models = (encoder, decoder)

# Define the loss function
reconstruction_loss_image = binary_crossentropy(K.reshape(inputs_image,(-1,im_size*im_size)), K.reshape(outputs[0],(-1,im_size*im_size)))
reconstruction_loss_bead = binary_crossentropy(K.reshape(inputs_bead,(-1,bead_size)), K.reshape(outputs[1],(-1,bead_size)))

reconstruction_loss_image *= im_size*im_size
reconstruction_loss_bead *= bead_size

reconstruction_loss = reconstruction_loss_image + reconstruction_loss_bead

kl_loss = 1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
kl_loss = K.sum(kl_loss, axis=-1)
kl_loss *= -0.5
vae_loss = K.mean(reconstruction_loss + kl_loss)
vae.add_loss(vae_loss)

# Compile it all
optimizer = keras.optimizers.Adam(lr=1e-3)
vae.compile(optimizer=optimizer)
```

### Training

```python
# train the autoencoder
epochs = 30
batch_size = 32
vae.fit([matrix, beads], epochs=epochs, batch_size=batch_size, shuffle=True)
```

Use the encoder to generate the mean values (`z_mean`) within the latent space of each encoded peptide.

```python
z_mean,z_logvar,z = encoder.predict([matrix,beads])
```

### Traversing the latent space
To help understand the latent space, I consider the peptides encountered when interpolating between some starting peptide ($\vec{start}$) and ending peptide ($\vec{end}$) from the mean values of the embedding. These points are choosen by selecting some maximum number of points $N$ in the line connecting $\vec{start}$ and $\vec{end}$.

$$\vec{point_n} = \vec{start} + t\ (\vec{end}-\vec{start})$$

Where $t \in [0,1]$ is evenly split into $N$ values generating all $\vec{point_n}$. The closest value in the latent space to each $\vec{point_n}$ in the embedding is then recorded as the peptide in the sequence producing the traversal from the starting peptide to the ending peptide.

```python
def interpolateSpace(startPeptide,endPeptide,numPoints=10):
    ''''
    Helper function for showing the peptides encountered when traversing between two points in the latent space. 
    # Inputs:
        startPeptide (string):
             string of length `WING_SIZE` of amino acid to start
        endPeptide (string):
             string of length `WING_SIZE` of amino acid to end
        numPoints (int):
             maximum number of peptides to show when interpolating between `startPeptide` and `endPeptide`   
    # Returns:
        none
    
    # Notes:
        Prints a string showing the peptides encountered when traversing between `startPeptide` and `endPeptide`
        with the number of MARTINI beads contained in each peptide in parentheses
    '''
    startPoint = z_mean[np.where(peptide[:,0]==startPeptide)[0][0]]
    endPoint = z_mean[np.where(peptide[:,0]==endPeptide)[0][0]]
    pointList = np.array([startPoint + t*(endPoint-startPoint) for t in np.linspace(0,1,numPoints+2)])
    peptideList = np.array([peptide[np.argmin(np.sum((z_mean-x0)**2,axis=1,keepdims=True))] for x0 in pointList])
    _,uniqueIndex = np.unique(peptideList[:,0],return_index=True)
    peptideList = peptideList[np.sort(uniqueIndex)]
    print(''.join(" {} ({}) -->".format(peptide[0].upper(),peptide[1]) for peptide in peptideList)[1:-4])
```

Peptide sequence between known good assembling peptides GAG and AVG.

```python
interpolateSpace('gag','avg')
```

## Visualize the latent space
Plot the 3D latent space `z` and color according to the number of beads per molecule. 

```python
import matplotlib.cm as cmx
from mpl_toolkits.mplot3d import Axes3D
def scatter3d(x,y,z, cs, colorsMap='jet'):
    cm = plt.get_cmap(colorsMap)
    cNorm = matplotlib.colors.Normalize(vmin=min(cs), vmax=max(cs))
    scalarMap = cmx.ScalarMappable(norm=cNorm, cmap=cm)
    fig = plt.figure()
    ax = Axes3D(fig)
    im = ax.scatter(x, y, z, c=scalarMap.to_rgba(cs), s=35)
    ax.set_xlabel(r'$z_1$')
    ax.set_ylabel(r'$z_2$')
    ax.set_zlabel(r'$z_3$')
    scalarMap.set_array(cs)
    
    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    
    cax = fig.add_axes([1.10, 0.05, 0.025, 0.95])
    cbar = fig.colorbar(scalarMap, orientation='vertical', cax=cax)
    cbar.ax.set_ylabel('no. of Martini beads', rotation=90)
    plt.savefig('VAE.pdf')
    plt.show()
```
